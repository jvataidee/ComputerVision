```python
import cv2
import argparse
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plt_imshow(title, image):
    # convert the image frame BGR to RGB color space and display it
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.axis('off')
    plt.show()
```

## **Drawing**

```python
canvas = np.zeros((300, 300, 3), dtype = 'uint8')
plt_imshow("Canvas", canvas)
```

```python
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
plt_imshow("Imagem", canvas)
```

```python
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red)
plt_imshow("Imagem", canvas)
```

```python
blue = (255, 0, 0)
cv2.rectangle(canvas, (10, 10), (60, 60), blue, 5)
plt_imshow("Imagem", canvas)
```

```python
cv2.rectangle(canvas, (50, 200), (200, 225), blue, -1)
plt_imshow("Imagem", canvas)
```

```python
cv2.circle(canvas, (250, 250), 10, red, -1)
plt_imshow("Imagem", canvas)
```

```python
canvas = np.zeros((300, 300, 3), dtype='uint8')
(cX, cY) = (canvas.shape[1]//2 , canvas.shape[0]//2)
white = (255, 255, 255)
```

```python
for radio in range(0, 175, 10):
    cv2.circle(canvas, (cX, cY), radio, white)

plt_imshow("circulos", canvas)
```
