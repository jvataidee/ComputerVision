```python
import cv2
import argparse
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def plt_imshow(title, image):
    # convert the image frame BGR to RGB color space and display it
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.axis('off')
    plt.show()
```

### Lighting Conditions

* high contrast: Maximize the contrast between the regions of interest in your image or objects
    * Generalizable: Your lighting conditions should be consistent enough that the work well from one object to the next
    * Stable: having stable, consisten and repeatable lighting conditions.
* Color Spaces:is just specific organization of color that allow us to consistently represent and reproduce colors
    * RGB: Red(0, 255) Green(0, 255) Blue(0, 255) -> 0 white 255 black (Cube)
    * HSV: Hue(Pure color, tons of red, (0, 179)) Saturation(White control, (0, 179)) Value(Control the lightness color,(0, 179))(cylinder rather)
    * l*a*b*: more especially is non-intuitive, Green -a, Blue -b,  Yellow +b, Red +a, White L, Black
    * Gray Scale: One channel tons of gray (0, 255)


```python
for (name, chan) in zip (("B", "G", "R"), cv2.split(image)):
    plt_imshow(name, chan)
```

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

for (name, chan) in zip (("H", "S", "V"), cv2.split(image2)):
    plt_imshow(name, chan)
```
